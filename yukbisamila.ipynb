```python
import pandas as pd
import numpy as np
from kmodes.kprototypes import KPrototypes
from sklearn.cluster import KMeans
```

```python
data = pd.read_excel('Data1.xlsx')
data
```

```python
data.drop('KECAMATAN', axis=1, inplace=True)
data
```

```python
data['STATUS HIV'] = data['STATUS HIV'].replace(['Tidak diketahui'], np.NaN)
data['HASIL TCM'] = data['HASIL TCM'].replace(['Tidak dilakukan'], np.NaN)
data['FOTO TORAKS'] = data['FOTO TORAKS'].replace(['Tidak dilakukan'], np.NaN)
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

data['JENIS KELAMIN'] = labelencoder.fit_transform(data['JENIS KELAMIN'])
#data['FOTO TORAKS'] = labelencoder.fit_transform(data['FOTO TORAKS'])
#data['STATUS HIV'] = labelencoder.fit_transform(data['STATUS HIV'])
data['FOTO TORAKS'].replace({ 'Negatif':0, 'Positif':1}, inplace=True)
data['STATUS HIV'].replace({'Negatif':0, 'Positif':1 }, inplace=True)
data['RIWAYAT DIABETES'] = labelencoder.fit_transform(data['RIWAYAT DIABETES'])
#data['HASIL TCM'] = labelencoder.fit_transform(data['HASIL TCM'])
data['HASIL TCM'].replace({'Rif resisten':0,'Rif Sensitif':1, 'Negatif':2}, inplace=True)
data['LOKASI ANATOMI (target/output)'] = labelencoder.fit_transform(data['LOKASI ANATOMI (target/output)'])
data
```

```python
encoded = pd.get_dummies(data['JENIS KELAMIN'], prefix='JENIS KELAMIN')
encoded.astype(int)
```

```python
data = data.drop(['JENIS KELAMIN'], axis = 1)
data['JENIS KELAMIN_0'], data['JENIS KELAMIN_1'] = encoded.astype(int)['JENIS KELAMIN_0'], encoded.astype(int)['JENIS KELAMIN_1']
#encoded.astype(int)
data
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
data = scaler.fit_transform(data)

data
```

```python
df = pd.DataFrame(data, columns =['UMUR', 'FOTO TORAKS', 'STATUS HIV', 'RIWAYAT DIABETES', 'HASIL TCM', 'LOKASI ANATOMI (target/output)', 'JENIS KELAMIN_0', 'JENIS KELAMIN_1'])
df
```

```python
df['JENIS KELAMIN_0'] = df['JENIS KELAMIN_0'].astype('object')
df['JENIS KELAMIN_1'] = df['JENIS KELAMIN_1'].astype('object')
df['FOTO TORAKS'] = df['FOTO TORAKS'].astype('object')
df['STATUS HIV'] = df['STATUS HIV'].astype('object')
df['RIWAYAT DIABETES'] = df['RIWAYAT DIABETES'].astype('object')
df['HASIL TCM'] = df['HASIL TCM'].astype('object')
df['LOKASI ANATOMI (target/output)'] = df['LOKASI ANATOMI (target/output)'].astype('object')
df['UMUR'] = df['UMUR'].astype('int')
print(df.dtypes)
```

```python
# melakukan k-means clustering pada atribut numerik
kmeans = KMeans(n_clusters=3, random_state=42).fit(df[['UMUR']])
df['cluster_umur'] = kmeans.labels_
```

```python
# melakukan k-modes clustering pada atribut kategorikal
from kmodes.kprototypes import KPrototypes

kp = KPrototypes(n_clusters=3, random_state=42).fit(df[['JENIS KELAMIN_0', 'JENIS KELAMIN_1','FOTO TORAKS', 'STATUS HIV', 'RIWAYAT DIABETES', 'HASIL TCM']])
df['cluster_kat'] = kp.labels_
```
